```python
df = pd.read_csv("ciudades.csv")
df["nombre"] = df["nombre"].apply(lambda x: x.lower())
df["nombre"]
```

```python
country = "Spain"
results = []  # Lista para almacenar los datos

for nombre in df["nombre"]:
    api_url = f"https://api.api-ninjas.com/v1/geocoding?city={nombre}&country={country}"
    response = requests.get(api_url, headers={'X-Api-Key': 'jHK0vdDDblcp5rqneJLIKQ==IwjVpsE9WGgay5XD'})
    
    if response.status_code == requests.codes.ok:
        data = response.json()  # Convertir la respuesta en JSON
        if data:  # Verificar si la respuesta contiene datos
            results.append(data[0])  # Agregar el primer resultado a la lista
            print(f"Datos obtenidos para {nombre}: {data[0]}")
    else:
        print("Error:", response.status_code, response.text)

# Convertir la lista de resultados en un DataFrame
df_results = pd.DataFrame(results)

# Mostrar el DataFrame resultante
print(df_results)
```

```python
def rellenar_states(filas_especificas, x, df_results):

    # Reemplazar NaN en la columna 'state' en las filas especificadas por el valor dado (x)
    df_results.loc[filas_especificas, 'state'] = df_results.loc[filas_especificas, 'state'].fillna(x)

    return df_results
```

```python
# CEUTA
filas_especificas = [149]  # Filas con NaN
nuevo_valor = 'Ceuta'        # El valor para reemplazar el NaN

# Ejecutar la función y obtener el DataFrame modificado
df_results = rellenar_states(filas_especificas, nuevo_valor, df_results)
```

```python
# Canary Islands
filas_especificas = np.r_[0:149, 150:638]  # Filas con NaN
nuevo_valor = 'Canary Islands'        # El valor para reemplazar el NaN

# Ejecutar la función y obtener el DataFrame modificado
df_results = rellenar_states(filas_especificas, nuevo_valor, df_results)
```
